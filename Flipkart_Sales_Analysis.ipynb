# Exploratory Data Analysis

## Gender

## Q1 plotting a bar chart for Gender and it's count?

```python
plt.figure(figsize=(10, 6), facecolor='yellow')

plt.title('Gender Count')

a = sns.countplot(y='Gender', hue='Gender', data=df, palette='Blues')

a.patch.set_facecolor('yellow')

for bars in a.containers:
    a.bar_label(bars)
```

# INSIGHT
+ The bar chart illustrates a gender distribution within the dataset, with females comprising a majority of 7832, compared to males, who account for 3407 individuals.

## Q2 plot a pie chart for total sales by gender

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum()

colors = sns.color_palette('Blues', n_colors=len(sales_gender))

plt.figure(figsize=(9, 6), facecolor='yellow')
plt.title('Total Sales By Gender')

patches, texts, autotexts = plt.pie(sales_gender['Amount'], labels=sales_gender['Gender'], autopct='', colors=colors, pctdistance=0.7)
plt.axis('equal')

female_sales = sales_gender[sales_gender['Gender'] == 'F']['Amount'].values[0]
male_sales = sales_gender[sales_gender['Gender'] == 'M']['Amount'].values[0]
plt.text(-0.4, 0, f"{female_sales}", fontsize=12, ha='center', va='center', color='black')
plt.text(0.4, 0, f"{male_sales}", fontsize=12, ha='center', va='top', color='black')

plt.legend(title='Gender', loc='upper left', bbox_to_anchor=(1, 0.5))
```

# INSIGHT
+ From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Q3 How do the counts vary across different age groups?

```python
plt.figure(figsize=(9, 6), facecolor='yellow')
plt.title('Age group wise count')

a = sns.countplot(data=df, x='Age Group', hue='Gender', palette='Blues')

a.patch.set_facecolor('yellow')
a.grid(False)

for bars in a.containers:
    a.bar_label(bars)
```

## Q5 What is the distribution of orders across top 10 states?

```python
plt.figure(figsize=(13, 6), facecolor='yellow')
plt.title('State-wise Distribution of Orders')

orders_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

a = sns.lineplot(data=orders_state, y='Orders', x='State', marker='o', color='blue')

a.patch.set_facecolor('yellow')
a.grid(True,color='gray')
for index, row in orders_state.iterrows():
    a.text(row['State'], row['Orders'], f'{row["Orders"]}', color='black', ha='left', va='center')
    
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.xticks(rotation=90)
plt.tight_layout()
```

## INSIGHT
+ Uttar Pradesh (UP) has the highest number of orders, with a total of 4807, indicating that it is the state with the highest demand or sales volume within the dataset.

## Q6 What are the Top 5 Customer by sales in different states?

```python
sales_by_customer = df.groupby(['State', 'Customer_name'])['Amount'].sum().reset_index()
sales_by_customer_sorted = sales_by_customer.groupby('State').apply(lambda x: x.nlargest(5, 'Amount')).reset_index(drop=True)

states = sales_by_customer_sorted['State'].unique()
num_states = len(states)

num_rows = (num_states + 2) // 3
fig, axes = plt.subplots(num_rows, 3, figsize=(12, 5*num_rows), sharex=True, facecolor='yellow')
fig.patch.set_facecolor('yellow')

colors = sns.color_palette('Blues', n_colors=5)

for i, state in enumerate(states):
    row = i // 3  
    col = i % 3   
    
    data = sales_by_customer_sorted[sales_by_customer_sorted['State'] == state]
    
    ax = sns.barplot(x='Amount', y='Customer_name', hue='Customer_name', data=data, ax=axes[row, col], palette=colors, dodge=False)
    
    ax.set_title(f'Top 5 Customers by Sales in {state}')
    ax.set_xlabel('Total Sales')
    ax.set_ylabel('Customer Name')
    ax.set_facecolor('yellow')
    
    if ax.legend_:
        ax.legend_.remove()

for i in range(num_states, num_rows*3):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
```

## INSIGHT
+ Analyzing the top 5 customers by sales across states offers a snapshot of regional buying patterns and customer loyalty. It reveals which customers contribute most significantly to sales in each area

## Q15 How does the gender distribution vary across different states?

```python
gender_distribution = df.groupby(['State', 'Gender']).size().reset_index(name='Count')

plt.figure(figsize=(12, 8), facecolor='yellow')
plt.title('Gender Distribution Across Different States')

ax = sns.barplot(data=gender_distribution, x='State', y='Count', hue='Gender', palette='Blues')

plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.grid(False)
plt.gca().patch.set_facecolor('yellow')
plt.legend(title='Gender', loc='upper right')
plt.xticks(rotation=90)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', 
                va='baseline', 
                xytext=(0, 5), 
                textcoords='offset points')
```

## INSIGHT

The insight from the chart indicates that across various states, there's a consistent trend of a higher proportion of female customers compared to male customers. This could suggest potential opportunities for targeted marketing strategies tailored towards female demographics in these states.

## Q16 Are there any noticeable patterns in the gender distribution based on product categories?

```python
gender_by_category = df.groupby(['Product_Category', 'Gender']).size().reset_index(name='Count')

plt.figure(figsize=(12, 8), facecolor='yellow')
plt.title('Gender Distribution by Product Category')

palette = sns.color_palette("Blues", n_colors=len(gender_by_category['Gender'].unique()))

ax = sns.barplot(data=gender_by_category, x='Product_Category', y='Count', hue='Gender', palette=palette)

plt.xlabel('Product Category')
plt.ylabel('Number of Customers')
plt.grid(False)
plt.gca().patch.set_facecolor('yellow')
plt.legend(title='Gender', loc='upper right')
plt.xticks(rotation=90)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', 
                va='baseline', 
                xytext=(0, 5), 
                textcoords='offset points')
```

## INSIGHT
The gender distribution across various product categories reveals a consistent trend of female dominance, with the majority of categories showing a higher proportion of female customers. This pattern is evident in product categories such as clothing, cosmetics, household items, and electronics. However, a notable exception is observed in the category of hand and power tools and books, where male customers are exclusively represented. This suggests a clear preference among male consumers for such products
